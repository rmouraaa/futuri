```python
#Importa√ß√µes
from openai_module import generate_prompt, ask_openai
from giphy_module import fetch_gifs
from user_interface import collect_responses, display_typewriter_effect
import re
from IPython.core.display import HTML
from IPython.core.display import clear_output

```

```python
class MIBYApp:
    def __init__(self):
        self.nome_widget = None
        self.idade_widget = None
        self.pergunta_widget = None

    def clean_response(self, response):
        # Remover emojis
        cleaned = re.sub(r'[\U00010000-\U0010ffff]', '', response)
        # Remover conte√∫do entre par√™nteses
        cleaned = re.sub(r'\(.*?\)', '', cleaned)
        return cleaned.strip()

    def extract_tag(self, response):
        # Procura por conte√∫do entre par√™nteses
        matches = re.findall(r'\(.*?\)', response)
        # Retorna a primeira correspond√™ncia ou None se n√£o houver correspond√™ncias
        return matches[0] if matches else None

    def on_submit_button_clicked(self, b):
        clear_output()
        print("Processando sua pergunta... Por favor, aguarde.")

        prompt = generate_prompt(self.nome_widget.value, self.idade_widget.value, self.pergunta_widget.value)
        raw_response = ask_openai(prompt)

        # Salvar resposta limpa na vari√°vel 'content'
        cleaned_response = self.clean_response(raw_response)

        # Extrair tag da resposta
        tag = self.extract_tag(raw_response)

        # Para exibi√ß√£o, remover apenas o conte√∫do entre par√™nteses
        display_response = re.sub(r'\(.*?\)', '', raw_response).strip()

        print(f"Resposta para {self.nome_widget.value}:")
        display_typewriter_effect(display_response)

        # Buscar e exibir GIFs relacionados
        if tag:
            gif_urls = fetch_gifs(tag)
            for url in gif_urls:
                display(HTML(f'<img src="{url}" alt="GIF relacionado">'))

    def on_retry_button_clicked(self, b):
        self.collect_user_input()

    def collect_user_input(self):
        self.nome_widget, self.idade_widget, self.pergunta_widget = collect_responses(
            self.on_submit_button_clicked,
            self.on_retry_button_clicked
        )

if __name__ == "__main__":
    app = MIBYApp()
    app.collect_user_input()
```
